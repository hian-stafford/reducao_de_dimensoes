# Principal Component Analysis (PCA)
Análise dos Principais Componentes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
%matplotlib inline
```

```python
# Plantas Iris
iris = datasets.load_iris()
```

```python
type(iris)
```

## Informações sobre o dataset das plantas Iris

```python
print(iris['DESCR'])
```

```python
X = iris['data']
y = iris['target']
```

```python
X.shape, y.shape
```

Como não é possível plotar estes dados num gráfico com 4 dimensões, é utilizado o PCA para solucionar este problema, diminuindo a quantidade de dimensões.

```python
X
```

Vamos normalizar estes dados, colocando num padrâo onde estes valores nâo fiquem destoando muito.

```python
X = StandardScaler().fit_transform(X)
```

```python
X
```

```python
iris['feature_names']
```

```python
pca = PCA(n_components=2)
```

```python
X.shape
```

### Principal Component (pc)

```python
pc = pca.fit_transform(X)
```

```python
pc.shape
```

```python
pc[:, 0]
```

```python
y
```

```python
pc[y == 2,0]
```

```python
plt.scatter(pc[:, 0], pc[:,1])
```

```python
plt.scatter(pc[y==0,0], pc[y==0,1], label ='Setosa')
plt.scatter(pc[y==1,0], pc[y==1,1], label ='Versicolour')
plt.scatter(pc[y==2,0], pc[y==2,1], label ='Virginica')

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
```

```python
pca.explained_variance_ratio_
```

```python
sum(pca.explained_variance_ratio_)
```

## Conclusão

O primeiro componente explica 72.9% dos dados e o segundo componente explica 22.8% dos dados, onde a soma explica 95.8% dos dados. Ou seja, perdemos apenas 4.2% quando simplificamos em 2 dimensões.
Vale a pena porque não perde muita informação, onde se torna possível visualizar em um gráfico 2D.

# Dígitos

```python
digitos = datasets.load_digits()
X = digitos['data']
y = digitos['target']
```

```python
X.shape, y.shape
```

```python
pca = PCA(n_components=2)
X_reduzido = pca.fit_transform(X)

print(f'Projetando {X.shape[1]} dimensões em apenas 2D')

plt.figure(figsize= (12,10))
plt.scatter(X_reduzido[:,0], X_reduzido[:,1], c=y,
            edgecolors='black', alpha=0.6, s=80,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.colorbar()
plt.show()
```

# TSNE

```python
from sklearn.manifold import TSNE
tsne =TSNE(random_state=123)

X_tsne = tsne.fit_transform(X)
```

```python
X_tsne.shape
```

```python
plt.figure(figsize=(12,10))
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=y,
            edgecolors='black', alpha=0.6, s=80,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.colorbar()
plt.show()
```

## Conclusão

TSNE é melhor para a visualização, mas é mais lento. Dependendo da quantidade dos dados, pode acabar não sendo tão viável utilizar este, e sim o PCA.

```python
X.shape
```

```python
pca = PCA(0.95)
```

```python
pca.fit_transform(X)
```

```python
pca.n_components_
```

Dessa forma, é possível escolher a quantidade de componentes/dimensões para manter a integridade dos dados. Neste exemplo, caso queira 90% dos dados, deve-se utilizar 21 componentes. 

É bom para processar mais rápido, podendo melhorar a qualidade de uma predição, caso seja necessário. Simplificar as dimensões pode ser algo positivo, porque a quantidade de dimensões faz com que o modelo fique mais complexo, sendo a melhor solução.

```python
np.arange(1, pca.n_components_+1)
```

```python
plt.figure(figsize=(8,4))
plt.bar(np.arange(1, pca.n_components_+1), pca.explained_variance_ratio_[:pca.n_components_],
        color = 'lightskyblue')
plt.xlabel('Componentes Principais')
plt.ylabel('Variação')
plt.xticks(np.arange(1, pca.n_components_+1))
plt.show()
```
